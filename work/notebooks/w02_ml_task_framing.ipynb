## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

A fixed rule (like "update pages older than 365 days") is too rigid. Content decay is a messy problem driven by many tangled, shifting signals. For example, the threshold for decay might be completely different for a "glossary page" versus a "news article," or depend heavily on the page's historical click-through rate. ML earns its place here because it can weigh these complex, multi-feature interactions simultaneously to prioritize the queue much more effectively than a human-written if-statement.

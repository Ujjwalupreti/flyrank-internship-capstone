## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/flyrank-bih/flyrank-ml-internship-starter/main/data/raw/content_refresh_anonymized.csv'
df = pd.read_csv(url)

# Calculate total pages and the exact class imbalance
total_pages = len(df)
declining_pages = df[df['trend_direction'] == 'down'].shape[0]
decline_rate = (declining_pages / total_pages) * 100

# Calculate a feature insight (e.g., age vs decline)
median_age_declining = df[df['trend_direction'] == 'down']['content_age_days'].median()
median_age_stable = df[df['trend_direction'] != 'down']['content_age_days'].median()

print(f"--- Dataset Reality Check ---")
print(f"1. Target Balance: Out of {total_pages:,} pages, {declining_pages:,} are actively declining ({decline_rate:.1f}%).")
print(f"2. Age Signal: The median age of a declining page is {median_age_declining} days, compared to {median_age_stable} days for stable/growing pages.")
```
